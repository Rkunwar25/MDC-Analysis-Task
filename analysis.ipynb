## ANALYSING YOUTUBE VIDEO TRENDS

1. FOR THE FILE CAVIDEOS - CANADIAN VIDEOS

```python
# Importing necessary libraries/packages : : 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sn
import warnings
warnings.filterwarnings("ignore")
```

```python
d=pd.read_csv("CAvideos.csv")
```

```python
a=d.columns
print("The columns are : ")
for i in a:
    print(i)
```

```python
rc=d.shape
print("Given dataset consists of",rc[0],"number of rows and",rc[1],"number of columns.")
```

```python
print("Description of different fields : ")
d.dtypes
```

```python
d.head()
```

## UNIVARIATE ANALYSIS

```python
d.describe()
```

```python
print("Number of videos trending during different dates : ")
d['trending_date'].value_counts()
```

```python
plt.figure(figsize=(10,8))

# Distribution of Views
plt.subplot(2, 2, 1)
plt.hist(d['views'], bins=50, color='skyblue')
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')

# Distribution of Likes
plt.subplot(2, 2, 2)
plt.hist(d['likes'], bins=50, color='lightgreen')
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')

# Distribution of Dislikes
plt.subplot(2, 2, 3)
plt.hist(d['dislikes'], bins=50, color='lightcoral')
plt.title('Distribution of Dislikes')
plt.xlabel('Dislikes')
plt.ylabel('Frequency')

# Distribution of Comment Count
plt.subplot(2, 2, 4)
plt.hist(d['comment_count'], bins=50, color='orange')
plt.title('Distribution of Comment Count')
plt.xlabel('Comment Count')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Analysis : Basing on the above graphs, we can visualise trends as follows:
1. For views column, majority of the videos fall into view count of about 0.0 to 0.2 hundred millions. It means the majority trending videos have views of about 10 to 20 millions.
2. For likes column, majority of the videos fall into view count of about less than 1 million.
3. Dislikes lie in range of less than 0.2 million indicating lesser the amount of dislikes, more the video trends.
4. Comments are also in range of less than 0.2 million.

All this indicates that newly released videos tend to TREND for a good amount of time and with time, as people view more and more they start to become less trendy.


Lets analyse further :

```python
corr_matrix = d[['views', 'likes', 'dislikes', 'comment_count']].corr()
plt.figure(figsize=(8,6))
sn.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

The above correlation map indicates relation between different categorical values.
we infer the following : 
1. More views indicate more likes, making video more trending.
2. More comments indicate more likes, another reason showing how people's interaction is almost proportional to whether they like a video or not.
3. There is a good diffrence between likes and dislikes, indicating people clearly either like a video or dislike a video; no in betweens.

```python
category_views = d.groupby('category_id')['views'].mean().sort_values(ascending=False)
category_likes = d.groupby('category_id')['likes'].mean().sort_values(ascending=False)
print("Most popular categories by average views:\n", category_views)
print("\nMost popular categories by average likes:\n", category_likes)
```

Inferences :
1. Category ID 10 has the highest average number of views at approximately 3.53 million.
2. Category ID 30 follows with an average of around 2.85 million views.
3. Category ID 29 has an average of about 1.56 million views.
4. Category ID 10 has the highest average number of likes at approximately 151,286 likes.
5. Category ID 29 follows with an average of around 138,414 likes.

This indicates that category 10 videos tend to have highest average views and likes and therefore videos in this category TREND more.

```python
from collections import Counter
tags = d['tags'].str.split('|').explode()
common_tags = Counter(tags).most_common(20)

tags, counts = zip(*common_tags)
plt.figure(figsize=(10,6))
plt.barh(tags, counts, color='purple')
plt.xlabel('Frequency')
plt.title('Top 20 Most Common Tags')
plt.gca().invert_yaxis()
plt.show()
```

The most viewership is on videos with tag "Funny" indicating more histerical videos are likely to be more trending.

## In conclusion, a video in Canada trends more with condition : 
1. It is newly released, and views are significantly low
2. More views and likes; whereas dislikes are less.
3. Most of the videos are of "funny" tagged.
